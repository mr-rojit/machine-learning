# <center>Normalization</center>
Normalization is one of the most frequently used data preparation techniques, which helps us to change the values of numeric columns in the dataset to use a common scale. <br>

Example :- If a person has a height of 6 feet, then "6" is its magniture and "feet" is its unit. <br>
           If a person has a weight of 75 kg, then "75" is its magniture and "kg" is its unit. <br>
          
As portrayed height and weight has its own unit and their magnitude has a huge difference. We can clearly see that
Weight has more advantage over height and our machine learning model will provide more emphasis on the weight column and this should not be the case. For this reason we normalise our our data in a "COMMON SCALE".



<h4 color='blue'>There are various types of normalization techniques</h4>

<li> Min-max scaler  - Mostly used </li>
<li> MaxAbs Scaling </li>
<li> Robust Scaling </li>


### MinMaxScaling
`Formula:` Xn = (X - Xminimum) / ( Xmaximum - Xminimum)  
By using this formula we can normalize every observation(column entry) to normal/comman scale between 0,1

```python
import pandas as pd
import seaborn as sns
```

```python
data = [
    [1, 6.1, 72],
    [2, 5.4, 61],
    [3, 5, 53],
    [4, 6.2, 74],
    [5, 5.1, 61],
    [6, 5.8, 70],
    [7, 5.5, 68],
    [8, 5.11, 70],   
]
df = pd.DataFrame(data=data, columns=['S.N', 'Height', 'Weight'])
df
```

### Ploting kde dist plot to see how the data looks before normalizing

```python
sns.kdeplot(df['Height'])
```

```python
sns.kdeplot(df['Weight'])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
columns_to_normalize = df[['Height', 'Weight']]


scaler.fit(columns_to_normalize)
normalized_data = scaler.transform(columns_to_normalize)
print(normalized_data)
```

#### We can see that maximum value of the colum is transformed into 1 and minimum to 0 -> of both columns

### <center>After normalizing</center>

```python
sns.kdeplot(normalized_data[:, 0], label='Height')
```

```python
sns.kdeplot(normalized_data[:, 1])
```

### Data's original shape is still persistent. Only the values are squished between 0 and 1 in the common scale
<b>Outliers in data also remains the same. </b>
